```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# File to Load
books_summary = "Resources/books_clean.csv"
# Read the modified GoodReads csv and store into Pandas DataFrame
# Hint: Build a DataFrame. You may need to use utf-8 encoding (encoding="utf-8")
books_summary_df = pd.read_csv(books_summary, encoding = "utf-8")
books_summary_df["Publication Year"] = books_summary_df["Publication Year"].values.astype(np.int64)
books_summary_df["Publication Year"] = books_summary_df["Publication Year"].replace(np.nan)
books_summary_df
```

```python
# Calculate the number of unique authors in the DataFrame
authors = books_summary_df["Authors"].nunique()
# Calculate the earliest and latest year a book was published
earliest = books_summary_df["Publication Year"].min()
latest = books_summary_df["Publication Year"].max()
# Calculate the total reviews column that contains the sum of reviews for each row
# Hint: use the pandas' sum(axis='columns') to get the sum across the columns 
books_summary_df["Review Total"] = books_summary_df.drop("Publication Year", axis = 1).sum(axis = 'columns')
total_reviews = books_summary_df["Review Total"].sum()
books_summary_df
earliest
```

```python
total_reviews
```

```python
# Place all of the data found into a summary DataFrame
data_summary_dict = {"Total Unique Authors" : authors,
                    "Earliest Year" : [earliest],
                    "Latest Year": latest,
                    "Total Reviews" : total_reviews}
# Display the summary data frame
data_summary_df = pd.DataFrame(data_summary_dict)
data_summary_df
```
